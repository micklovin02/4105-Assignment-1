```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load the breast cancer dataset
breast = load_breast_cancer()
X = breast.data
Y = breast.target

# Convert to DataFrame
breast_input = pd.DataFrame(X)

# Split the dataset into training and test sets (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Scale and standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Problem 1: Logistic Regression
# (i) Logistic Regression without weight penalty

# Train the logistic regression model
logreg = LogisticRegression()
logreg.fit(X_train_scaled, y_train)

# Predictions
y_pred_logreg = logreg.predict(X_test_scaled)

# Calculate evaluation metrics
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
precision_logreg = precision_score(y_test, y_pred_logreg)
recall_logreg = recall_score(y_test, y_pred_logreg)
f1_logreg = f1_score(y_test, y_pred_logreg)

# Confusion Matrix
conf_matrix_logreg = confusion_matrix(y_test, y_pred_logreg)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
plt.imshow(conf_matrix_logreg, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix for Logistic Regression')
plt.colorbar()
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.yticks([0, 1], ['Benign', 'Malignant'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Report results
print("(i) Logistic Regression without weight penalty:")
print("Accuracy:", accuracy_logreg)
print("Precision:", precision_logreg)
print("Recall:", recall_logreg)
print("F1 Score:", f1_logreg)
```

```python
# (ii) Logistic Regression with weight penalty
# We'll use L2 penalty (Ridge regularization)

# Train the logistic regression model with L2 penalty
logreg_penalty = LogisticRegression(penalty='l2')
logreg_penalty.fit(X_train_scaled, y_train)

# Predictions
y_pred_logreg_penalty = logreg_penalty.predict(X_test_scaled)

# Calculate evaluation metrics
accuracy_logreg_penalty = accuracy_score(y_test, y_pred_logreg_penalty)
precision_logreg_penalty = precision_score(y_test, y_pred_logreg_penalty)
recall_logreg_penalty = recall_score(y_test, y_pred_logreg_penalty)
f1_logreg_penalty = f1_score(y_test, y_pred_logreg_penalty)

# Report results
print("\n(ii) Logistic Regression with weight penalty:")
print("Accuracy:", accuracy_logreg_penalty)
print("Precision:", precision_logreg_penalty)
print("Recall:", recall_logreg_penalty)
print("F1 Score:", f1_logreg_penalty)

```

```python
# Problem 2: Naive Bayesian Model
# Train the naive Bayesian model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Predictions
y_pred_nb = nb_model.predict(X_test)

# Calculate evaluation metrics
accuracy_nb = accuracy_score(y_test, y_pred_nb)
precision_nb = precision_score(y_test, y_pred_nb)
recall_nb = recall_score(y_test, y_pred_nb)
f1_nb = f1_score(y_test, y_pred_nb)

# Report results
print("\n(i) Naive Bayesian Model:")
print("Accuracy:", accuracy_nb)
print("Precision:", precision_nb)
print("Recall:", recall_nb)
print("F1 Score:", f1_nb)

```

```python
# (ii) Plot Classification Metrics for Logistic Regression and Naive Bayesian Model
metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
logreg_metrics = [accuracy_logreg, precision_logreg, recall_logreg, f1_logreg]
logreg_penalty_metrics = [accuracy_logreg_penalty, precision_logreg_penalty, recall_logreg_penalty, f1_logreg_penalty]
nb_metrics = [accuracy_nb, precision_nb, recall_nb, f1_nb]

plt.figure(figsize=(10, 6))
plt.bar(metrics_names, logreg_metrics, width=0.2, label='Logistic Regression')
plt.bar(metrics_names, logreg_penalty_metrics, width=0.2, label='Logistic Regression with Penalty', align='edge')
plt.bar(metrics_names, nb_metrics, width=0.2, label='Naive Bayesian', align='center')
plt.xlabel('Metrics')
plt.ylabel('Score')
plt.title('Comparison of Classification Metrics')
plt.legend()
plt.show()

```

```python
# Problem 3: Logistic Regression with PCA
# Perform PCA and identify the optimal number of principal components

# Number of components to consider
num_components = range(1, 31)
accuracy_scores = []

for n in num_components:
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    # Train the logistic regression model with PCA components
    logreg_pca = LogisticRegression()
    logreg_pca.fit(X_train_pca, y_train)

    # Predictions
    y_pred_pca = logreg_pca.predict(X_test_pca)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred_pca)
    accuracy_scores.append(accuracy)

# Plot accuracy scores over different numbers of components
plt.plot(num_components, accuracy_scores, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Number of Principal Components')
plt.show()

# Optimal number of components
optimal_components = num_components[np.argmax(accuracy_scores)]
print("Optimal number of principal components:", optimal_components)

```

```python
# Problem 4: Bayesian Classifier with PCA
# Repeat Problem 3 using Bayesian classifier instead of logistic regression

# Train the Bayesian classifier with PCA components
bayes_pca = GaussianNB()
bayes_pca.fit(X_train_pca, y_train)

# Predictions
y_pred_bayes_pca = bayes_pca.predict(X_test_pca)

# Calculate evaluation metrics
accuracy_bayes_pca = accuracy_score(y_test, y_pred_bayes_pca)
precision_bayes_pca = precision_score(y_test, y_pred_bayes_pca)
recall_bayes_pca = recall_score(y_test, y_pred_bayes_pca)
f1_bayes_pca = f1_score(y_test, y_pred_bayes_pca)

# Report results
print("\nBayesian Classifier with PCA:")
print("Accuracy:", accuracy_bayes_pca)
print("Precision:", precision_bayes_pca)
print("Recall:", recall_bayes_pca)
print("F1 Score:", f1_bayes_pca)
```
